```python
import tensorflow as tf
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
data=load_breast_cancer()
```

```python
scaler=StandardScaler()
X_train,X_test,y_train,y_test=train_test_split(scaler.fit_transform(data.data),data.target,test_size=0.33,random_state=42)

X_train.shape,X_test.shape
```

```python
N,D=X_train.shape
```

```python
model=tf.keras.Sequential([tf.keras.layers.Input(shape=(D,)),tf.keras.layers.Dense(1,activation="sigmoid")])
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
r=model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(r.history['loss'],label="Train set loss")
plt.plot(r.history['val_loss'],label="Test set loss")
plt.title("Loss of the model in epoches")
plt.legend()
plt.show()
```

```python
plt.plot(r.history['accuracy'],label="Train set accuracy")
plt.plot(r.history['val_accuracy'],label="Test set accuracy")
plt.title("Accuracy of the model in epoches")
plt.legend()
plt.show()
```
